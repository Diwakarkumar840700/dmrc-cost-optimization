```python
input_file = '/content/train-metadata.csv'

# Count the number of columns in the header
with open(input_file, 'r') as file:
    header = file.readline()
    num_columns = len(header.split(','))

# Print lines with incorrect number of columns
with open(input_file, 'r') as file:
    lines = file.readlines()

for i, line in enumerate(lines):
    if len(line.split(',')) != num_columns:
        print(f"Line {i+1} has an incorrect number of columns: {line}")
```
